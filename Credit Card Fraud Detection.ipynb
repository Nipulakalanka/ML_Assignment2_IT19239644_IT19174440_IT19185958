```python
#This dataset highly unbalanced
# 0 ---> legit transaction
# 1 ---> fradulent transaction

# Separating the data for analysis
legit_transcation = raw_data[raw_data.Class == 0]
fradulent_transaction = raw_data[raw_data.Class ==1]
```

```python
#Under Sampling(build a sample dataset contatining simillar distribution of legit transactions and fradulent transactions)
#Random Sampling
legit_sample = legit_transcation.sample(n=492) 
```

```python
#Concatenating two Dataframes
new_rawdata=pd.concat([legit_sample,fradulent_transaction],axis=0)
```

```python
new_rawdata.head()
```

```python
new_rawdata.tail()
```

```python
# 0 ---> legit transaction
# 1 ---> fradulent transaction
new_rawdata['Class'].value_counts()
```

```python
new_rawdata.groupby('Class').mean()
```

```python
#splitting the data into features and targets
x=new_rawdata.drop(columns='Class',axis=1)
y=new_rawdata['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#Split the data into training data and tesing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#Model Training
#Logistic Regression
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train, y_train)
```

```python
#accuray on training data model.score method
training_data_accuracy = model.score(x_train, y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
#accuray on test data model.score method
test_data_accuracy = model.score(x_test,y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```
